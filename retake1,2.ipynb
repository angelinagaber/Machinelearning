```python
TrainData_m =TrainData_y.size
TrainData_m
```

```python
TrainData_x = np.concatenate([np.ones((TrainData_m,1)), TrainData_x], axis =1)
TrainData_x
```

```python
def computeCostMulti(TrainData_x, TrainData_y, theta):
    
    
    TrainData_m = TrainData_y.shape[0] #number of training examples  
    
    
    j=0
    
    j=np.dot((np.dot(TrainData_x,theta)-TrainData_y),(np.dot(TrainData_x,theta)-TrainData_y))/(2*TrainData_m) 
    
    return j
```

```python
def gradientDescentMulti(TrainData_x,TrainData_y, theta, alpha, num_iters):
    
    TrainData_m = TrainData_y.shape[0] 
    
    theta = theta.copy()
    j_history= []
    for i in range(num_iters):
        sumofh0x=np.dot(TrainData_x,theta)
        theta =theta-((alpha/TrainData_m)*(np.dot(TrainData_x.T,sumofh0x- TrainData_y)))
        
        j_history.append(computeCostMulti(TrainData_x, TrainData_y, theta))
    return theta, j_history
```

```python

alpha = 0.03
num_iters = 100

theta = np.zeros(3) 
theta, j_history = gradientDescentMulti(TrainData_x, TrainData_y, theta, alpha, num_iters)  
pyplot.plot(np.arange(len(j_history)), j_history, lw=2)
pyplot.xlabel('number of itierations')
pyplot.ylabel('Cost j')
```

```python
def computeCostMulti2(TrainData_x, TrainData_y, theta):

 TrainData_m = TrainData_y.shape[0]
 j=0
 j=np.dot((np.dot(np.square(TrainData_x),theta)-TrainData_y),(np.dot(np.square(TrainData_x),theta)-TrainData_y))/(2*TrainData_m)
 return j
```

```python
def gradientDescentMulti2(TrainData_x, TrainData_y, theta, alpha, num_iters):
    TrainData_m = TrainData_y.shape[0]
    theta=theta.copy()
    j_history = []
    for i in range(num_iters):
        sumofh0x=np.dot(np.square(TrainData_x),theta)
        theta = theta-((alpha/TrainData_m)*(np.dot(np.square(TrainData_x).T,sumofh0x -TrainData_y)))
        j_history.append(computeCostMulti2(TrainData_x, TrainData_y, theta))
     
    return theta, j_history
```

```python
alpha=0.01
num_iters = 100
theta = np.zeros(3)
theta, j_history = gradientDescentMulti2(TrainData_x, TrainData_y, theta, alpha, num_iters)
pyplot.plot(np.arange(len(j_history)), j_history, lw=2)
pyplot.xlabel('number of itirations')
pyplot.ylabel('cost j')
```

```python
def computeCostMulti3(TrainData_x, TrainData_y, theta):

 TrainData_m = TrainData_y.shape[0]  
 j=0
    
 j =np.dot((np.dot(np.power(TrainData_x,3),theta)-TrainData_y),(np.dot(np.power(TrainData_x,3),theta)-TrainData_y))/(2*TrainData_m)
 return j
```

```python
def gradientDescentMulti3(TrainData_x, TrainData_y, theta, alpha, num_iters):
    TrainData_m = TrainData_y.shape[0]
    theta=theta.copy()
    j_history = []
    for i in range(num_iters):
        sumofh0x=np.dot(np.power(TrainData_x,3),theta)
        theta = theta-((alpha/TrainData_m)*(np.dot(np.power(TrainData_x,3).T,sumofh0x -TrainData_y)))
        j_history.append(computeCostMulti3(TrainData_x, TrainData_y, theta))
     
    return theta, j_history
```
